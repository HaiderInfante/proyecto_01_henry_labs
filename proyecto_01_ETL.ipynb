## Visualización del archivo steam_games

Realizar la extracción de los datos del archivo steam_games y se almacena en un dataframe para una debida manipulación y transformación.

```python
# Utilizamos la libreria gzip para abrir el archivo comprimido
with gzip.open('02_Datasets_PI MLOps_STEAM/steam_games.json.gz', 'rt', encoding='utf-8') as steam_games:
    #Leemos cada linea del archivo
    data_steam_games = steam_games.readlines()
# Itera sobre cada linea leida quitando los espacios en blanco y caracteres de nueva linea al final de cada elemento de la lista  
data_steam_games = map(lambda x: x.rstrip(), data_steam_games)
# Se crea una cadena json válida concatenando todos los elementos de la lista
data_steam_games_json = "[" + ','.join(data_steam_games) + "]"
# Se cargan los datos en formato json en un dataframe de pandas
df_steam_games = pd.read_json(data_steam_games_json)
# Se genera el encabezado del dataframe
df_steam_games
```

## Visualización del archivo users_items

creamos una nueva columna con la fecha de posted en formato YYYY-MM-DD

```python
# Definir una función para extraer y convertir las fechas
def extract_and_convert_date(date_str):
    match = re.search(r'(\w+) (\d+), (\d+)', date_str)
    if match:
        month = match.group(1)
        day = match.group(2)
        year = match.group(3)
        try:
            # Intentar construir una fecha válida y luego formatearla
            formatted_date = pd.to_datetime(f"{year}-{month}-{day}", format='%Y-%B-%d', errors='coerce')
            return formatted_date.strftime('%Y-%m-%d')
        except ValueError:
            # Si la conversión falla, devolver NaN
            return None
    else:
        # Si no se encuentra un patrón válido, devolver NaN
        return None

# Aplicar la función a la columna 'Fecha' y crear una nueva columna 'Nueva_Fecha'
df_second_function_file_one['new_date_format'] = df_second_function_file_one['posted'].apply(extract_and_convert_date)

df_second_function_file_one
```

```python
df_second_function_file_one = df_second_function_file_one.dropna(subset=['new_date_format'])
df_second_function_file_one
```

Realizar el analisis de sentimientos aplicando NLP

```python
# Descargar el recurso VADER
nltk.download('vader_lexicon')

# Inicializar el analizador de sentimientos de VADER
sia = SentimentIntensityAnalyzer()

# Función para clasificar las reseñas
def classify_sentiment(review):
    sentiment_scores = sia.polarity_scores(review)
    compound_score = sentiment_scores['compound']
    if compound_score >= 0.05:
        return 2  # Positivo
    elif compound_score <= -0.05:
        return 0  # Malo
    else:
        return 1  # Neutral

# Aplicar la función para clasificar las reseñas y crear una nueva columna
df_sixth_function['review'] = df_sixth_function['review'].apply(classify_sentiment)

# Imprimir el DataFrame resultante
df_sixth_function
```
